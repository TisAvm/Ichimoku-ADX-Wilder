```python
# Ichimoku-ADX-Wilder Strategy Backtesting on Minute-Level Data
# This notebook performs comprehensive backtesting with M2M calculations
# Trading session: 9:20 AM to 3:25 PM on minute-level data

import sys
import os
import warnings
warnings.filterwarnings('ignore')

# Add current directory to path for imports
current_dir = os.path.dirname(os.path.abspath('__file__'))
sys.path.append(current_dir)

# Import our backtesting modules
from signal_generator import SignalGenerator
from backtesting import IchimokuADXBacktester, run_complete_backtest

print("ğŸš€ Ichimoku-ADX-Wilder Backtesting System Initialized")
print("=" * 60)
print("ğŸ“Š Features:")
print("  â€¢ Minute-level data backtesting")
print("  â€¢ Trading session: 9:20 AM - 3:25 PM")
print("  â€¢ Mark-to-Market (M2M) calculations")
print("  â€¢ Comprehensive performance metrics")
print("  â€¢ Multiple scenario analysis")
print("  â€¢ Transaction cost modeling")
print("=" * 60)
```

```python
# Step 1: Load and Analyze Signals Data
print("ğŸ“ˆ STEP 1: Loading and Analyzing Signals Data")
print("-" * 50)

# Initialize signal generator with our signals file
signals_file = "data/ichimoku_adx_wilder_signals.csv"

try:
    signal_gen = SignalGenerator(signals_file)
    
    # Get date range of available signals
    start_date, end_date = signal_gen.get_date_range()
    print(f"âœ… Signals loaded successfully!")
    print(f"ğŸ“… Available date range: {start_date} to {end_date}")
    
    # Analyze signal quality
    print("\nğŸ” Analyzing signal quality...")
    quality_analysis = signal_gen.analyze_signal_quality()
    
    if quality_analysis:
        print("\nğŸ“Š SIGNAL QUALITY METRICS:")
        for key, value in quality_analysis.items():
            if isinstance(value, dict) and key == 'adx_stats':
                print(f"\nğŸ¯ ADX Statistics:")
                for sub_key, sub_value in value.items():
                    if isinstance(sub_value, float):
                        print(f"   {sub_key}: {sub_value:.2f}")
                    else:
                        print(f"   {sub_key}: {sub_value}")
            elif not isinstance(value, dict):
                if isinstance(value, float):
                    print(f"   {key}: {value:.2f}")
                else:
                    print(f"   {key}: {value}")
    
except Exception as e:
    print(f"âŒ Error loading signals: {e}")
    print("Please ensure the signals file exists at 'data/ichimoku_adx_wilder_signals.csv'")
    raise
```

```python
# Step 2: Run Single Backtest Scenario
print("\nğŸ¯ STEP 2: Running Single Backtest Scenario")
print("-" * 50)

# Configuration for backtesting
config = {
    'initial_capital': 100000,  # â‚¹1,00,000
    'position_size': 0.1,       # 10% of capital per trade
    'symbol': 'NIFTY',
    'scenario_name': 'Standard Backtest'
}

print(f"âš™ï¸  BACKTEST CONFIGURATION:")
print(f"   ğŸ’° Initial Capital: â‚¹{config['initial_capital']:,}")
print(f"   ğŸ“ Position Size: {config['position_size']*100}% of capital")
print(f"   ğŸ“Š Symbol: {config['symbol']}")
print(f"   â° Trading Hours: 9:20 AM - 3:25 PM")
print(f"   ğŸ“ˆ Data Frequency: Minute-level")

try:
    # Run the backtest
    print(f"\nğŸš€ Running backtest scenario: {config['scenario_name']}")
    
    scenario_result = signal_gen.run_backtest_scenario(
        scenario_name=config['scenario_name'],
        start_date=start_date,
        end_date=end_date,
        initial_capital=config['initial_capital'],
        position_size=config['position_size'],
        symbol=config['symbol']
    )
    
    if 'error' not in scenario_result:
        print("âœ… Backtest completed successfully!")
        
        # Display key metrics
        metrics = scenario_result['metrics']
        print(f"\nğŸ“Š KEY PERFORMANCE METRICS:")
        print(f"   ğŸ’¹ Total Return: {metrics['Total Return (%)']:.2f}%")
        print(f"   ğŸ“ˆ Final Portfolio Value: â‚¹{metrics['Final Portfolio Value']:,.2f}")
        print(f"   ğŸ“‰ Maximum Drawdown: {metrics['Maximum Drawdown (%)']:.2f}%")
        print(f"   ğŸ¯ Win Rate: {metrics['Win Rate (%)']:.2f}%")
        print(f"   ğŸ”¢ Total Trades: {metrics['Total Trades']}")
        print(f"   âš¡ Sharpe Ratio: {metrics['Sharpe Ratio']:.2f}")
        
    else:
        print(f"âŒ Backtest failed: {scenario_result['error']}")
        
except Exception as e:
    print(f"âŒ Error running backtest: {e}")
    # Continue with the notebook even if this fails
```

```python
# Step 3: Multiple Scenario Analysis
print("\nğŸ”¬ STEP 3: Multiple Scenario Analysis")
print("-" * 50)

# Run multiple scenarios with different parameters
print("ğŸ”„ Running multiple backtesting scenarios...")
print("This will test different position sizes and capital amounts")

try:
    scenarios = signal_gen.run_multiple_scenarios()
    
    print(f"\nâœ… Completed {len([s for s in scenarios if 'error' not in s])} scenarios successfully!")
    
    # Compare scenarios
    signal_gen.compare_scenarios(scenarios)
    
    print("\nğŸ† SCENARIO WINNERS:")
    
    # Find best scenarios by different metrics
    valid_scenarios = [s for s in scenarios if 'error' not in s]
    
    if valid_scenarios:
        # Best return
        best_return = max(valid_scenarios, key=lambda x: x['metrics']['Total Return (%)'])
        print(f"   ğŸ“ˆ Highest Return: {best_return['scenario_name']} ({best_return['metrics']['Total Return (%)']:.2f}%)")
        
        # Best Sharpe ratio
        best_sharpe = max(valid_scenarios, key=lambda x: x['metrics']['Sharpe Ratio'])
        print(f"   âš¡ Best Sharpe Ratio: {best_sharpe['scenario_name']} ({best_sharpe['metrics']['Sharpe Ratio']:.2f})")
        
        # Lowest drawdown
        best_drawdown = max(valid_scenarios, key=lambda x: x['metrics']['Maximum Drawdown (%)'])  # Less negative is better
        print(f"   ğŸ›¡ï¸  Lowest Drawdown: {best_drawdown['scenario_name']} ({best_drawdown['metrics']['Maximum Drawdown (%)']:.2f}%)")
        
        # Highest win rate
        best_winrate = max(valid_scenarios, key=lambda x: x['metrics']['Win Rate (%)'])
        print(f"   ğŸ¯ Highest Win Rate: {best_winrate['scenario_name']} ({best_winrate['metrics']['Win Rate (%)']:.2f}%)")
    
except Exception as e:
    print(f"âŒ Error in scenario analysis: {e}")
    # Continue even if this fails
```

```python
# Step 4: Detailed M2M Analysis and Recent Signals
print("\nğŸ’¹ STEP 4: Mark-to-Market Analysis & Recent Signals")
print("-" * 50)

try:
    # Generate recent signals for live trading reference
    print("ğŸ“Š Generating recent trading signals (last 30 days)...")
    recent_signals = signal_gen.generate_trading_signals_for_live(lookback_days=30)
    
    if not recent_signals.empty:
        print(f"\nâœ… Found {len(recent_signals)} actionable signals in the last 30 days")
        
        # Display latest signals
        latest_signals = recent_signals.tail(10)
        print(f"\nğŸ• LATEST 10 ACTIONABLE SIGNALS:")
        
        for idx, signal in latest_signals.iterrows():
            signal_strength = "ğŸ”¥ Strong" if abs(signal['total_signal']) >= 3 else ("âš¡ Medium" if abs(signal['total_signal']) >= 2 else "ğŸ’¡ Weak")
            direction = "ğŸ“ˆ BUY" if signal['signal_type'] == 'BUY' else "ğŸ“‰ SELL"
            
            print(f"   {signal['datetime'].strftime('%Y-%m-%d %H:%M')} | {direction} | Price: â‚¹{signal['close']:.2f} | {signal_strength} | ADX: {signal.get('adx', 'N/A')}")
    
    # M2M Analysis from the best performing scenario
    if 'scenario_result' in locals() and 'error' not in scenario_result:
        print(f"\nğŸ’° MARK-TO-MARKET (M2M) ANALYSIS:")
        print(f"   Based on scenario: {scenario_result['scenario_name']}")
        
        backtester = scenario_result['backtester']
        
        if backtester.backtest_results is not None:
            # Calculate daily M2M
            backtest_data = backtester.backtest_results
            
            # Get daily portfolio values
            daily_m2m = backtest_data.groupby(backtest_data['timestamp'].dt.date).agg({
                'portfolio_value': ['first', 'last', 'min', 'max'],
                'unrealized_pnl': 'last'
            }).round(2)
            
            daily_m2m.columns = ['Day_Start', 'Day_End', 'Day_Low', 'Day_High', 'Unrealized_PnL']
            daily_m2m['Daily_PnL'] = daily_m2m['Day_End'] - daily_m2m['Day_Start']
            daily_m2m['Daily_Return_%'] = (daily_m2m['Daily_PnL'] / daily_m2m['Day_Start'] * 100).round(2)
            
            print(f"\nğŸ“ˆ DAILY M2M SUMMARY (Last 10 Days):")
            latest_m2m = daily_m2m.tail(10)
            
            for date, row in latest_m2m.iterrows():
                pnl_indicator = "ğŸ“ˆ" if row['Daily_PnL'] >= 0 else "ğŸ“‰"
                print(f"   {date} | {pnl_indicator} P&L: â‚¹{row['Daily_PnL']:,.2f} ({row['Daily_Return_%']:+.2f}%) | Portfolio: â‚¹{row['Day_End']:,.2f}")
            
            # Overall M2M statistics
            total_days = len(daily_m2m)
            profitable_days = (daily_m2m['Daily_PnL'] > 0).sum()
            avg_daily_return = daily_m2m['Daily_Return_%'].mean()
            best_day = daily_m2m['Daily_PnL'].max()
            worst_day = daily_m2m['Daily_PnL'].min()
            
            print(f"\nğŸ“Š M2M STATISTICS:")
            print(f"   ğŸ“… Total Trading Days: {total_days}")
            print(f"   ğŸ“ˆ Profitable Days: {profitable_days} ({profitable_days/total_days*100:.1f}%)")
            print(f"   ğŸ“Š Average Daily Return: {avg_daily_return:.2f}%")
            print(f"   ğŸ† Best Day P&L: â‚¹{best_day:,.2f}")
            print(f"   ğŸ“‰ Worst Day P&L: â‚¹{worst_day:,.2f}")
    
except Exception as e:
    print(f"âŒ Error in M2M analysis: {e}")

print(f"\nğŸ‰ BACKTESTING ANALYSIS COMPLETED!")
print("=" * 60)
print("ğŸ“ Check the './results/' directory for detailed output files:")
print("   ğŸ“„ backtest_results.csv - Complete minute-by-minute data")
print("   ğŸ’¼ trades.csv - Individual trade records")
print("   ğŸ“Š metrics.csv - Performance metrics")
print("   ğŸ”„ scenario_comparison.csv - Scenario comparison")
print("   ğŸ• recent_signals.csv - Recent actionable signals")
print("=" * 60)
```
